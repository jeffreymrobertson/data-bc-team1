```python
#Dependencies and Setup
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from scipy.stats import linregress
```

```python
# Reading the fashion data file
fashion_df = pd.read_csv("Resources/mock_fashion_data_uk_us.csv")
# Display the data table for preview
fashion_df.head()
```

```python
# Checking the number of products.
num_product = fashion_df["Product Name"].nunique()
num_product
```

```python
#Counting number of brands
brand_count = fashion_df["Brand"].value_counts()
brand_count
```

```python
# Created a Pandas bar plot for Brand count
plot_pandas = brand_count.plot.bar(color='tab:blue')  
# Setting up the xlabel, ylabel, and title 
plt.xlabel("Brand")
plt.ylabel("Number of Brands")
plt.title("Brand Count")
```

```python
# Generating a bar plot showing the total number of rows 
x_axis = brand_count.index.values
y_axis = brand_count.values

# Creating a Pyplot bar plot
plt.bar(x_axis, y_axis, color='tab:blue', alpha=0.8, align='center')

# Setting the xlabel and ylabel, title 
plt.xlabel("Brand")
plt.ylabel("Number of Brands")
plt.title("Brand Count")
plt.xticks(rotation="vertical")

plt.show()
```

## Pie chart representing Brand vs Value Count
 

```python
size_data = fashion_df["Brand"].value_counts()
plt.title("Brand vs Value Count")
size_data.plot.pie(autopct= "%1.1f%%")
plt.show()
```

## Pie chart representing Brand vs Price
 

```python
fashion_df.groupby(['Brand']).sum().plot(kind='pie', y='Price', autopct='%1.0f%%',
                               figsize=(11, 11),title='Brand Vs Price')
```

## Bar chart representing Brand vs Customer Reviews vs Social Media Comments vs      Feedback


```python
columns_selected=['Brand','Customer Reviews','Social Media Comments','feedback']
```

```python
# countplot of each categorical column
fig, axis = plt.subplots(nrows=2, ncols=2, figsize=(25, 10))

index = 0
for row in range(2):
    for col in range(2):
        sns.countplot(data=fashion_df, x=columns_selected[index], ax=axis[row, col])
        index += 1
        
plt.show()
```

## Pie chart representing Age Vs Social Media Comments
 

```python
fashion_df.groupby(['Social Media Comments']).sum().plot(kind='pie', y='Age', autopct='%1.0f%%',
                               figsize=(8, 8),title='Age Vs Social Media Comments')
```

## Pie chart representing Brand Vs Ratings

```python
fashion_df.groupby(['Brand']).sum().plot(kind='pie', y='Rating', autopct='%1.0f%%',
                               figsize=(11, 11),title='Brand Vs Rating')
```
