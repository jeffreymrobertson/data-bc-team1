```python
#Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Load data

fashion_df = pd.read_csv("Resources/mock_fashion_data_uk_us.csv")
fashion_df
```

```python
#Summarize statistics of numerical data

fashion_df.describe()
```

```python
#Create new DataFrame with only Style Attributes and Available Sizes columns

style_size_df=fashion_df[["Style Attributes","Available Sizes"]]
style_size_df.head()
```

```python
#List the unique Style Attributes from the dataset

style_size_df["Style Attributes"].unique()
```

```python
#List the unique Available Sizes from the dataset

style_size_df["Available Sizes"].unique()
```

```python
#Count the total of each size

style_size_df['Available Sizes'].value_counts()
```

```python
#Create data frame to group by Style Attributes

grouped_style_df=style_size_df.groupby("Style Attributes")
grouped_style_df.head()
```

```python
# Count the number of each size by Style Attributes

size_counts_style=grouped_style_df['Available Sizes'].value_counts()
size_counts_style
```

```python
#Create DataFrame with counts of each size by Style Attribute

size_counts_style_df=pd.DataFrame(size_counts_style)
size_counts_style_df.head()
```

```python
#Create pivot table to summarize counts of each size by Style Attributes

pivot_table1 = pd.pivot_table(size_counts_style_df, values="count", index="Style Attributes", columns="Available Sizes", aggfunc="sum")
print(pivot_table1)
```

```python
#Plot an unstacked bar chart 

figure1=pivot_table1.plot(kind='bar')
plt.xlabel("Style")
plt.ylabel("Count")
plt.title("Count of Sizes by Style Attributes")
```

```python
# Plot a stacked bar chart

figure2=pivot_table1.plot(kind='bar',stacked=True)

plt.xlabel("Style")
plt.ylabel("Count")
plt.title("Count of Sizes by Style Attributes")
```

```python
#Create new DataFrame to show Brand and Size Available

brand_size_df=fashion_df[["Brand","Available Sizes"]]
brand_size_df.head()
```

```python
#Create data frame to group by Brand

grouped_brand_df=brand_size_df.groupby("Brand")
grouped_brand_df.head()
```

```python
# Count the number of each size by Brand

size_counts_brand=grouped_brand_df['Available Sizes'].value_counts()
size_counts_brand
```

```python
#Create DataFrame with counts of each size by Brand

size_counts_brand_df=pd.DataFrame(size_counts_brand)
size_counts_brand_df.head()
```

```python
#Create pivot table to summarize counts of each size by Brand

pivot_table2 = pd.pivot_table(size_counts_brand_df, values="count", index="Brand", columns="Available Sizes", aggfunc="sum")
print(pivot_table2)
```

```python
# Plot a stacked bar chart

figure3=pivot_table2.plot(kind='bar',stacked=True)

plt.xlabel("Brand")
plt.ylabel("Count")
plt.title("Count of Sizes by Brand")
```
